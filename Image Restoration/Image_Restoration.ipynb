```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from skimage import restoration
from math import sqrt
```

```python
image = cv2.imread('lena.png', 0)
cv2_imshow(image)
```

```python
image_2 = cv2.GaussianBlur(image, (7,7), 0)
cv2_imshow(image_2)
```

```python
def RMSE(image, image_2):
  errors = []
  N = image.shape[0] * image.shape[1]
  sum = 0
  for i in range(len(image)):
    for j in range(len(image[i])):
      error = image[i][j] - image_2[i][j]
      sum += error
      errors.append(error)
  square_root = sqrt(sum)
  rmse = square_root/N
  return rmse
```

```python
rmse = RMSE(image, image_2)
print(rmse);
```
